```python
import pandas as pd
import sklearn
from sklearn.preprocessing import Imputer
```

```python
data=pd.read_csv('daily_weather.csv')
data
#print (data.columns) doubt*
```

```python
data[data.isnull().any(axis='columns')]#for each row
```

```python
del data['number']
```

```python
imp = Imputer(missing_values='NaN', strategy='mean', axis=0, copy=False)#to impute along columns
df = imp.fit_transform(data)
df = pd.DataFrame(data)
df.columns = data.columns
data = df
data
#fillna vs imputer D
#data[data.isnull().any(axis='columns')]
```

```python
# Convert to a Classification Task.Binarize the relative_humidity_3pm to 0 or 1.
clean_data = data.copy()
clean_data['high_humidity_label'] = (clean_data['relative_humidity_3pm'] > 24.99)*1
print(clean_data['high_humidity_label'])
```

```python
y=clean_data[['high_humidity_label']].copy()
y
```

```python
clean_data['relative_humidity_3pm'].head()
```

```python
y.head()
```

```python
morning_features = ['air_pressure_9am','air_temp_9am','avg_wind_direction_9am','avg_wind_speed_9am',
        'max_wind_direction_9am','max_wind_speed_9am','rain_accumulation_9am',
        'rain_duration_9am']
X = clean_data[morning_features].copy()
```

```python
X.columns
```

```python
y.columns
```

```python
##d
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
#type(X_train)
#type(X_test)
#y_train.describe()
#y_test
X_train[X_train.isnull().any(axis=1)]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
humidity_classifier = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)
humidity_classifier.fit(X_train, y_train)
```

```python
type(humidity_classifier)
```

```python
predictions = humidity_classifier.predict(X_test)
```

```python
predictions[:10]
```

```python
y_test['high_humidity_label'][:10]
```

```python
accuracy_score(y_true = y_test, y_pred = predictions)
```
